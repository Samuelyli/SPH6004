```python
def print_size(y_train, y_test):
    print(str(len(y_train[y_train['label']==1])) + '(' + str(round(len(y_train[y_train['label']==1])/len(y_train)*100,1)) + '%)' + ' non-survivors in training set')
    print(str(len(y_train[y_train['label']==0])) + '(' + str(round(len(y_train[y_train['label']==0])/len(y_train)*100,1)) + '%)' + ' survivors in training set')
    print(str(len(y_test[y_test['label']==1])) + '(' + str(round(len(y_test[y_test['label']==1])/len(y_test)*100,1)) + '%)' + ' non-survivors in test set')
    print(str(len(y_test[y_test['label']==0])) + '(' + str(round(len(y_test[y_test['label']==0])/len(y_test)*100,1)) + '%)' + ' survivors in test set')
```

```python
perc_class1 = 0.4

print('Before balancing')
print_size(y_train, y_test)

if perc_class1 > 0:
        
    class0_indices = y_train[y_train['label'] == 0].index
    class1_indices = y_train[y_train['label'] == 1].index

    class0_size = round(np.int((len(y_train[y_train['label'] == 1])*(1 - perc_class1)) / perc_class1),0)
    
    np.random.seed(10)
        
    random_indices = np.random.choice(class0_indices, class0_size, replace=False)    
        
    X_train = pd.concat([X_train.loc[random_indices],X_train.loc[class1_indices]])
    y_train = pd.concat([y_train.loc[random_indices],y_train.loc[class1_indices]])

    print('After balancing')
    print_size(y_train, y_test)
```

```python
#Define Performance
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix

def performance(y, y_pred, print_ = 1, *args):   

    fpr, tpr, thresholds = metrics.roc_curve(y, y_pred, pos_label=1)

    AUC = metrics.auc(fpr, tpr)

    if args:
        threshold = args[0]
    else:
        difference = abs((1-fpr) - tpr)
        threshold = thresholds[difference.argmin()]        
        
    y_pred[y_pred >= threshold] = 1
    y_pred[y_pred < threshold] = 0
    
    tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()
    sensitivity = tp/(tp+fn)
    specificity = tn/(tn+fp)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    
    if print_ == 1:
        print('Threshold: ' + str(round(threshold,2)))
        print('TP: ' + str(tp))
        print('TN: ' + str(tn))
        print('FP: ' + str(fp))
        print('FN: ' + str(fn))
        print("Accuracy: " + str( round(accuracy, 2 )))
        print('Sensitivity: ' + str(round(sensitivity,2)))
        print('Specificity: ' + str(round(specificity,2)))                
        print('AUC: ' + str(round(AUC,2)))
    
        plt.figure(figsize = (8,6))
        plt.scatter(x = fpr, y = tpr, label = None)
        plt.plot(fpr, tpr, label = 'Classifier', zorder = 1)
        plt.plot([0, 1], [0, 1], 'k--', label = 'Random classifier')
        plt.scatter(x = 1 - specificity, y = sensitivity, c = 'black', label = 'Operating point', zorder = 2)
        plt.legend()
        plt.xlabel('1 - specificity')
        plt.ylabel('sensitivity')
        plt.show()
        

    return threshold, AUC, sensitivity, specificity
```

```python
from sklearn.model_selection import cross_val_score, KFold, cross_validate
from sklearn.feature_selection import RFE, f_regression
from sklearn.linear_model import (LinearRegression, Ridge, Lasso, ElasticNet)

reg = ElasticNet (random_state=0, alpha=0.001, l1_ratio=0.5)
reg.fit(X_train_norm, y_train.values.ravel())

print(reg.coef_)
```

```python
def rfe_select_featurs(X_train_norm, y_train, estimator, num_features) -> List[str]:
    rfe_selector = RFE(estimator=estimator, 
                       n_features_to_select=num_features, 
                       step=10, verbose=5)
    rfe_selector.fit(X_train_norm, y_train)
    rfe_support = rfe_selector.get_support()
    rfe_feature = X_train_norm.loc[:,rfe_support].columns.tolist()
    print(str(len(rfe_feature)), 'selected features')
    
    return rfe_feature
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from statistics import mean

cross_val_scores_ridge = []

alpha = []

for i in range(1, 9):
	ridgeModel = Ridge(alpha = i * 0.25)
	ridgeModel.fit(X_train_norm, y_train)
	scores = cross_val_score(ridgeModel, X_train_norm, y_train, cv = 10)
	avg_cross_val_score = mean(scores)*100
	cross_val_scores_ridge.append(avg_cross_val_score)
	alpha.append(i * 0.25)

for i in range(0, len(alpha)):
	print(str(alpha[i])+' : '+str(cross_val_scores_ridge[i]))
```

```python
def abc_evaluation(abc, X_train_norm, y_train, X_test_norm, y_test, print_):
    
    print('TRAINING SET')
    y_pred_prob_train = abc.predict_proba(X_train_norm)
    threshold, AUC_train, sens_train, spec_train = performance(y_train, np.delete(y_pred_prob_train, 0, 1), print_)

    print('TEST SET')
    y_pred_prob_test = abc.predict_proba(X_test_norm)
    _, AUC_test, sens_test, spec_test = performance(y_test, np.delete(y_pred_prob_test, 0, 1), print_, threshold)
    
    results_train = pd.DataFrame(data = [[threshold, AUC_train, sens_train, spec_train, X_train_norm.shape[1]]],
                           columns = ['Threshold','AUC', 'Sensitivity', 'Specificity', '# features'])

    results_test = pd.DataFrame(data = [[threshold, AUC_test, sens_test, spec_test, X_train_norm.shape[1]]],
                           columns = ['Threshold','AUC', 'Sensitivity', 'Specificity', '# features'])
        
    return results_train, results_test, y_pred_prob_train, y_pred_prob_test
```
